```python
import bokeh
print(bokeh.__version__)
```

```python
# Import necessary libraries
from bokeh.plotting import figure
from bokeh.io import output_file, show, output_notebook, push_notebook
from bokeh.models import ColumnDataSource, HoverTool, CategoricalColorMapper
from bokeh.layouts import row, column, gridplot
from bokeh.models.widgets import Tabs, Panel

output_notebook()
```

```python
from bokeh.plotting import figure, show

f = figure()
f.line([10, 20, 38, 40, 50], [160, 70, 0, 40, 50], line_width=12)
show(f)
```

```python
#Circle Glyph
p = figure(x_axis_label="x axis", y_axis_label="y axis", tools="pan,box_zoom")

p.circle(x=[50, 30, 40], y=[10, 20, 30], size=25, color="blue", alpha=0.7, line_color="yellow")

show(p)
```

```python
# Square Glyph
t = figure(plot_width=200, plot_height=500)
t.square([10, 12, 13, 14, 15], [16, 17, 12, 14, 15], size=26, color="blue", alpha=0.7)
show(t)
```

```python
# Multiple Lines
y = figure(plot_width=500, plot_height=500)
y.multi_line(xs=[[12, 13, 12], [31, 41, 16]], ys=[[11, 12, 42], [14, 721, 181]], color=["red", "blue"], alpha=[0.8, 0.3], line_width=5)
show(y)
```

```python
# Wedge Glyph
z = figure(width=500, height=500)
z.wedge(x=[1, 2, 3], y=[1, 2, 3], radius=0.2, start_angle=-4, end_angle=5.2, color="black", alpha=0.7, direction="clock")
show(z)
```

```python
# annular Wedge Glyph
q = figure(width=500, height=500)
q.annular_wedge(x=[10, 28, 30], y=[10, 20, 30], inner_radius=0.5, outer_radius=0.75, start_angle=-5, end_angle=5.2, color="blue")
show(q)
```

```python
# Combine Circle and Line on the same plot

x = [18, 20, 30, 40]
y = [60, 75, 80, 90]

u = figure(width=400, height=500)

u.line(x, y, line_width=2)
u.circle(x, y, fill_color="black", size=9)
show(u)
```

```python
# Bar Plot
g1 = figure(title="Bar Plot")

x = (1, 2, 3, 4, 9)
y = (1, 2, 3, 4, 5)

g1.hbar(x, right=y, height=0.5)

show(g1)
```

```python
# Vertical Bar Plot

p = figure(title="Vertical Bar Plot")

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p.vbar(x=x, top=y, width=0.5)
show(p)
```

```python
import random
# Scatter Plot
g3 = figure(title='Scatter Plot')

x = [random.random() for _ in range(126)]
y = [random.random() + 2 for _ in range(126)]

g3.scatter(x, y)

show(g3)
```

```python
# Area Plot
import numpy as np

x = [10, 20, 30, 40, 50]
y1 = np.array([2, 4, 7, 8, 9])
y2 = np.array([1, 2, 3, 8, 4])

h = figure(width=400, height=400)

h.varea(x, y1, y2=y2, fill_color="blue")

show(h)
```

```python
# Patch Plot
w = figure(width=400, height=400)
w.patch(x=[1, 4, 6, 9], y=[2, 3, 7, 8], color="navy")
show(w)
```
